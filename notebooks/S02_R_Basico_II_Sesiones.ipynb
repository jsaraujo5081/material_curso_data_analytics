## Creación, acceso y modificación de diccionarios

Los diccionarios son estructuras de datos a partir de vectores atómicos y listas que permiten almacenar información en pares de *clave* y *valor*, facilitando la organización y recuperación eficiente de datos. A diferencia de los vectores y listas, donde los elementos se acceden mediante posiciones numéricas, los diccionarios permiten acceder a la información a través de identificadores específicos (claves), lo que los convierte en herramientas especialmente útiles para representar datos estructurados y relaciones entre elementos. En **R**, los diccionarios son ampliamente utilizados en tareas de análisis y ciencia de datos, ya que permiten configurar parámetros eficientemente y modelar de información compleja dada su flexibilidad y rapidez de acceso.

## Definición y uso de funciones para automatización de procesos

Las funciones constituyen bloques de código reutilizables diseñados para ejecutar tareas específicas dentro de un programa. Su principal objetivo es modularizar la solución de un problema, permitiendo dividir procesos complejos en partes más pequeñas, organizadas y fáciles de mantener. Mediante el uso de funciones, es posible reducir la repetición de código, mejorar la legibilidad de los programas y facilitar la detección de errores. En **R**, las funciones pueden recibir parámetros de entrada y retornar resultados, lo que favorece la creación de programas más dinámicos, reutilizables y escalables. Vale mencionar que su uso te permitirá desarrollar procesos de manera más eficiente en distintos ámbitos de la preparación, análisis  y modelamiento de datos.

### Ejercicio 3

Escribe una función que devuelva True (Verdadero) si un número entero cualquiera es par, y False en caso contrario.

```r
es_par <- function (n) {
    par = n %% 2 == 0
    return(par)
}
```

## Creación y uso de bucles o estructuras repetitivas

Los bucles, también conocidos como estructuras repetitivas, permiten ejecutar un conjunto de instrucciones múltiples veces de manera automática. Estas estructuras son fundamentales cuando se requiere procesar colecciones de datos, realizar tareas iterativas o automatizar operaciones repetitivas. El uso adecuado de bucles contribuye significativamente a la eficiencia y simplicidad de un programa, evitando la necesidad de escribir instrucciones redundantes. En **R**, los bucles más utilizados son los llamados `for` y `while`, cada uno orientado a diferentes tipos de procesos iterativos. Por una parte los bucles `for` facilitan el realizar tareas específicas en cada uno de los elementos de objetos iterables (i.e vectores, listas o diccionarios); mientras que los bucles `while` se especializan en repetir procedimientos mientras se cumpla una condición específica.

### Ejercicio 1

Utiliza un bucle `for` para transformar el diccionario de clientes en otro diccionario a modo de lista nombrada `cliente_list` donde cada elemento contenga tenga como denominación el nombre del cliente y la edad de cada persona sea su valor. 

```r
# Definir lista vacía
cliente_list <- list()

# Ejecutar el bucle en cada elemento del diccionario
for (nombre in names(cliente_dicc)) {
    cliente_list[nombre] = cliente_dicc[nombre]
}
cliente_list
```

Con la lista obtenida y utilizando un bucle `for`, imprime para cada cliente un mensaje que diga lo siguiente:

*El cliente de nombre (aquí va el nombre del cliente) tiene (aquí va su edad) años.*

```r
for (cliente in names(cliente_list)) {
    print(paste("El cliente de nombre", cliente, "tiene", cliente_list[cliente], "años."))
}
```

### Ejercicio 3

Utiliza un bucle `while` para recorrer la lista de clientes y mostrar la información del primer jubilado (mayor de 65 años) registrado. 

```r
# Definir variable para evaluar condición
edad_prueba <- 0

# Definir variable para iterar sobre la lista de clientes
pos <- 1

# Generar bucle hasta encontrar primer cliente jubilado
while (edad_prueba <= 65) {
    edad_prueba = cliente_list[[pos]]
    pos = pos + 1
}

print(paste(
    "El cliente de nombre",
    names(cliente_list)[pos - 1],
    "tiene",
    cliente_list[[pos - 1]], 
    "años."
))
```

### Ejercicio 4

Utiliza bucles for y while para crear un vector atómico simple `clientes_rep` que contenga cada nombre de cliente repetido 3 veces.

```r
# Crear vector vacio para guardar resultados
clientes_rep <- c()

# Ejecutar bucle for que recorra lista de clientes
for (cliente in names(cliente_list)) {
    
    # Definir iterable para bucle while
    rep <- 1

    # Ejecutar bucle while que repita el nombre 3 veces
    while (rep <= 3) {
        clientes_rep = c(clientes_rep, cliente)
        rep = rep + 1
    }

}

print(clientes_rep)
```

## Creación y uso de estructuras condicionales

Las estructuras condicionales permiten que un programa tome decisiones y ejecute diferentes acciones según se cumplan determinadas condiciones lógicas. Este mecanismo constituye la base del control de flujo en programación, ya que posibilita adaptar el comportamiento del programa frente a distintos escenarios o entradas de información. A través de condicionales, un algoritmo puede evaluar expresiones lógicas y seleccionar la ruta de ejecución más adecuada. En **R**, las instrucciones `if` y `else` proporcionan una forma clara y flexible de implementar procesos de decisión dentro de un programa.

### Ejercicio 1

Recorre el diccionario de clientes y extrae en un vector atómico `clientes_jovenes` todos los nombres de los clientes que tengan menos de 45 años.

```r
# Crear vector vacío para almacenar nombres de clientes jovenes
clientes_jovenes <- c()

# Definir bucle for para recorrer el diccionario
for (nombre in names(cliente_dicc)) {

    edad_cliente = unname(cliente_dicc[nombre])

    # Guardar nombre en lista si cumple con condicion mediante un if
    if (edad_cliente < 45) {
        clientes_jovenes = c(clientes_jovenes, nombre)
    }
        
}

print(clientes_jovenes)
```

### Ejercicio 2

Toma como referencia el código del ejercicio anterior de forma que ahora se guarden las edades de los clientes jóvenes en una lista llamada `edades_jovenes`, y en otra lista `edades_viejos` se guarden las edades de los clientes que no son jóvenes.

```r
# Crear vectores vacios para almacenar edades de clientes jovenes y "no jovenes"
edades_jovenes <- c()
edades_viejos <- c()

# Definir bucle for para recorrer el diccionario
for (nombre in names(cliente_dicc)) {

    edad_cliente = unname(cliente_dicc[nombre])

    if (edad_cliente < 45) {
        edades_jovenes = c(edades_jovenes, edad_cliente)
    } else {
        edades_viejos = c(edades_viejos, edad_cliente)
    }
        
}

print(edades_jovenes)
print(edades_viejos)
```

Calcula ahora el promedio de edades de clientes jóvenes y viejos e imprime los resultados con un mensaje adecuado. Utiliza para esto las funciones `sum` y `length`.

```r
prom_jovenes <- sum(edades_jovenes) / length(edades_jovenes)
prom_viejos <- sum(edades_viejos) / length(edades_viejos)

print(paste(
    "El promedio de edad de clientes jóvenes es de ",
    formatC(prom_jovenes, format = "f", digits = 1),
    " años, mientras que el de clientes viejos es de ",
    formatC(prom_viejos, format = "f", digits = 1),
    ".",
    sep = ""
))
```

### Ejercicio 3

Aleph busca generar campañas específicas de mercadeo para clientes menores a 30 años, otra para clientes entre 30 y 45 años, y una tercera para clientes cercanos a la jubilación (de igual o más de 45 años). Modifica el código generado en el ejercicio anterior a fin de que el mismo permita contar con vectores atómicos de nombres de clientes en cada uno de los segmentos relevantes.

```r
# Crear vectores vacoos para almacenar resultados de edades
nombres_sub30 <- c()
nombres_sub45 <- c()
nombres_over45 <- c()

# Definir bucle para recorrer el diccionario
for (nombre in names(cliente_dicc)) {

    edad_cliente = unname(cliente_dicc[nombre])

    if (edad_cliente < 30) {
        nombres_sub30 = c(nombres_sub30, nombre)
    } else if (edad_cliente < 45) {
        nombres_sub45 = c(nombres_sub45, nombre)
    } else {
        nombres_over45 = c(nombres_over45, nombre)
    }
        
}

# Crear lista nombrada consolidada
mercados_list <- list(
    "Mercado 1" = nombres_sub30,
    "Mercado 2" = nombres_sub45,
    "Mercado 3" = nombres_over45
)
mercados_list
```
