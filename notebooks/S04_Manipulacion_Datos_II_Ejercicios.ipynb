## Unión de dataframes

### Ejercicio 1

Considera ahora los siguientes dataframes de ejemplo:

```r
dataset_1 <- tibble(
    'key' = c('K0','K1','K2','K3'),
    'A' = c('A0', 'A1', 'A2', 'A3'),
    'B' = c('B0', 'B1', 'B2', 'B3')
)

dataset_2 = tibble(
    'key' = c('K0','K1','K2','K3'),
    'A' = c('A4', 'A5', 'A6', 'A7'),
    'B' = c('B4','B5','B6','B7'),
)
```

Cárgalos e imprímelos en notebook.

````r
# Cargar datasets
dataset_1 <- tibble(
    'key' = c('K0','K1','K2','K3'),
    'A' = c('A0', 'A1', 'A2', 'A3'),
    'B' = c('B0', 'B1', 'B2', 'B3')
)

dataset_2 = tibble(
    'key' = c('K0','K1','K2','K3'),
    'A' = c('A4', 'A5', 'A6', 'A7'),
    'B' = c('B4','B5','B6','B7'),
)
````

````r
# Imprimir datasets
print(dataset_1)
print(dataset_2)
````

Usa la función `inner_join` para unir los datasets  y usando la columna key como clave.

````r
resultado <- dataset_1 %>%
inner_join(
    dataset_2,
    by = "key"
)
print(resultado)
````

Realiza la misma unión que en el caso anterior, solo que ahora especifica sufijos distintos para columnas con los mismos nombres en ambos datasets.

````r
resultado <- dataset_1 %>%
inner_join(
    dataset_2,
    by = "key",
    suffix = c("_d1", "_d2")
)
print(resultado)
````

### Ejercicio 2

Considera los siguientes dataframes de ejemplo:

```r
dataset_3 = tibble(
    'key1' = c('K0', 'K0', 'K1', 'K2'),
    'key2' = c('K0', 'K1', 'K0', 'K1'),
    'A' = c('A0', 'A1', 'A2', 'A3'),
    'B' = c('B0', 'B1', 'B2', 'B3')
)

dataset_4 = tibble(
    'key1' = c('K0', 'K1', 'K1', 'K2'),
    'key2' = c('K0', 'K0', 'K0', 'K0'),
    'C' = c('C0', 'C1', 'C2', 'C3'),
    'D' = c('D0', 'D1', 'D2', 'D3')
)
```

Cárgalos e imprímelos en pantalla.

````r
# Cargar datos
dataset_3 = tibble(
    'key1' = c('K0', 'K0', 'K1', 'K2'),
    'key2' = c('K0', 'K1', 'K0', 'K1'),
    'A' = c('A0', 'A1', 'A2', 'A3'),
    'B' = c('B0', 'B1', 'B2', 'B3')
)

dataset_4 = tibble(
    'key1' = c('K0', 'K1', 'K1', 'K2'),
    'key2' = c('K0', 'K0', 'K0', 'K0'),
    'C' = c('C0', 'C1', 'C2', 'C3'),
    'D' = c('D0', 'D1', 'D2', 'D3')
)
````

````r
# Imprimir datos
print(dataset_3)
print(dataset_4)
````

Une los datasets con un *inner join*, usando las llaves key1 y key2 a la vez.

````r
resultado <- dataset_3 %>%
inner_join(
    dataset_4,
    by = c("key1", "key2")
)
print(resultado)
````

Repite la unión anterior pero ahora utiliza un *full join*.

````r
resultado <- dataset_3 %>%
full_join(
    dataset_4,
    by = c("key1", "key2")
)
print(resultado)
````

Ahora usa un *right join*.

````r
resultado <- dataset_3 %>%
right_join(
    dataset_4,
    by = c("key1", "key2")
)
print(resultado)
````

## Aplicación de funciones para enriquecer datos

### Ejercicio 1

Ejecuta el siguiente codigo para crear un dataset de forma aleatoria.

```r
set.seed(123)

# Definir fechas cualquiera
fechas = date("2023-01-01") + 0:99

# Simular datos
ventas <- sample(100:700, size = 100)
gastos_marketing <- sample(50:200, size = 100)
gastos_personal <- sample(80:300, size = 100)

# Crear dataframe
df <- tibble(
    fecha = fechas,
    ventas = ventas,
    gastos_marketing = gastos_marketing,
    gastos_personal = gastos_personal

df %>% sample_n(10)
```

````r
set.seed(123)

# Definir fechas cualquiera
fechas = date("2023-01-01") + 0:99

# Simular datos
ventas <- sample(100:700, size = 100)
gastos_marketing <- sample(50:200, size = 100)
gastos_personal <- sample(80:300, size = 100)

# Crear dataframe
df <- tibble(
    fecha = fechas,
    ventas = ventas,
    gastos_marketing = gastos_marketing,
    gastos_personal = gastos_personal
)

df %>% sample_n(10)
````

Crea una nueva columna estado en la que indique si "hay ganancia" (ventas >= gastos_marketing + gastos_personal) o "hay pérdida".

````r
df <- df %>%
mutate(estado = if_else(
    ventas > gastos_marketing + gastos_personal,
    "hay ganancia",
    "hay perdida"
))

df %>% sample_n(10)
````

### Ejercicio 2

Crea una nueva columna llamada razon_perdida en la cual se especifique "marketing" si los gastos de este tipo son mayores a los de personal o "personal" en caso contrario, pero solamente cuando el estado sea de pérdida.

````r
df <- df %>%
mutate(razon_perdida = if_else(
    estado == "hay ganancia",
    "no aplica",
    if_else(
        gastos_marketing > gastos_personal,
        "marketing",
        "personal"
    )
))

df %>% sample_n(10)
````

### Ejercicio 3

Si las ventas superan el valor de 500 en un día se tiene un buen día. Crea una columna de tipo buleano que refleje este hecho.

````r
df <- df %>%
mutate(es_dia_bueno = if_else(
    ventas > 500,
    TRUE,
    FALSE
))

df %>% sample_n(10)
````

## Visualizaciones mediante gráficos

### Ejercicio 1

Tomando como referencia el dataset del apartado anterior, crea un gráfico de linea con la evolución de ventas y que tenga varios atributos de forma para hacerlo presentable. Añade además una línea horizonal en el valor de 500 para distinguir los días "buenos".

````r
options(repr.plot.width = 12, repr.plot.height = 6)
ggplot(
    data = df,
    mapping = aes(
        x = fecha,
        y = ventas
    )
) +
geom_line(
    linewidth = 1,
    color = "dodgerblue"
) +
geom_point(
    size = 3,
    color = "darkblue"
) +
geom_hline(
    yintercept = 500,
    color = "red",
    linewidth = 1.5,
    linetype = "dashed"
) +
labs(
    title = "Evolución de ventas",
    x = "",
    y = "USD"
) +
theme_classic() +
theme(
    panel.grid.major.y = element_line(color = "black", linetype = "dashed"),
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.text = element_text(size = 15),
    axis.title = element_text(size = 15)
)
````

### Ejercicio 2

Crea un histograma que muestre y compare la distribución de gastos en marketing y los gastos de personal. Recuerda incluir atributos de forma.

````r
options(repr.plot.width = 10, repr.plot.height = 6)
ggplot(data = df) +
geom_histogram(
    mapping = aes(
        x = gastos_marketing,
        fill = "marketing"
    ),
    #fill = "goldenrod",
    color = "goldenrod",
    alpha = 0.5,
    bins = 10
) +
geom_histogram(
    mapping = aes(
        x = gastos_personal,
        fill = "personal"
    ),
    #fill = "green",
    color = "green4",
    alpha = 0.5,
    bins = 10
) +
scale_fill_manual(values = c("goldenrod", "green4")) +
labs(
    title = "Distribucion de gastos",
    x = "USD",
    y = "",
    fill = ""
) +
theme_classic() +
theme(
    panel.grid.major.y = element_line(color = "black", linetype = "dotted"),
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.text = element_text(size = 15),
    axis.title = element_text(size = 15),
    legend.position = "top",
    legend.text = element_text(size = 15)
)
````

### Ejercicio 3

Crea un gráfico de cajón para tener una idea general de la distribución, dispersión y valores atipicos en las ventas, gastos de marketing y gastos de personal. Recuerda incluir atributos de forma.

````r
options(repr.plot.width = 8, repr.plot.height = 8)
ggplot(data = df) +
geom_boxplot(
    mapping = aes(
        x = ventas,
        y = 0
    ),
    fill = "lightblue",
    width = 0.5
) +
geom_boxplot(
    mapping = aes(
        x = gastos_marketing,
        y = 1
    ),
    fill = "pink",
    width = 0.5
) +
geom_boxplot(
    mapping = aes(
        x = gastos_personal,
        y = 2
    ),
    fill = "pink",
    width = 0.5
) +
scale_y_continuous(
    breaks = c(0, 1, 2), 
    labels = c("ventas", "gastos marketing", "gastos personal")
) +
labs(
    title = "Distribucion de ventas y gastos",
    x = "USD",
    y = ""
) +
theme_classic() +
theme(
    panel.grid.major.x = element_line(color = "black", linetype = "dotted"),
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.text = element_text(size = 15),
    axis.title = element_text(size = 15)
)
````
