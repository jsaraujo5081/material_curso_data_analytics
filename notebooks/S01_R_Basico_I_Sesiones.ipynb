````r
tryCatch(
    expr = {
        nombre <- "Aleph"
        activos <- 2000
        pasivos <- "1500"
        patrimonio <- activos - pasivos
        razon_end <- pasivos / patrimonio
        print(paste("El patrimonio de", nombre, "es de", patrimonio,"USD, y su razon de endeudamiento es de", razon_end, sep = " "))
    },
    error = function (e) {
        print("Existe algún error")
    }
)
````

## Manipulación y procesamiento de cadenas de texto

Las cadenas de texto representan secuencias de caracteres utilizadas para almacenar y procesar información textual. Su manipulación es esencial en múltiples aplicaciones, como el análisis de datos, la validación de información, la automatización de procesos y la interacción con usuarios. El manejo de cadenas incluye operaciones como concatenación, búsqueda, reemplazo, extracción y transformación de texto. Gracias a las herramientas incorporadas en **R**, es posible realizar este tipo de tareas de forma sencilla y eficiente, facilitando el procesamiento de información escrita.

### Ejercicio 1

John intentó enviar un mensaje de no más de 50 caracteres a su amigo Max para decirle que le había enviado bitcoins de regalo. Pero se equivocó al escribir el mensaje y el programa arrojó un error.

```r
message <- 'Hey pal. I've sent you some bitcoins. Check your wallet!'
print(message)
```

````r
message <- 'Hey pal. I\'ve sent you some bitcoins. Check your wallet!'
print(message)
````

¿Se ajusta el mensaje a la longitud de los caracteres? Compruébalo mediante la función `nchar`.

````r
length <- nchar(message)
print(length <= 50)
````

Una vez corregido y enviado el mensaje. Max respondió a John, pero el mensaje venía con algunos problemas:

* Caracteres "\#" entre todas las palabras.
* Espacios tanto al principio como al final del mensaje.
* Letras en mayúsculas y minúsculas de forma aleatoria.

```r
message <- "  hEY#bRO,#I#nEed#YouR#hElP#bUYiNg#bTc#tONigHt!     "
```

Ayuda a John a entender el mensaje.

````r
# Definir el mensaje original
message <- "  hEY#bRO,#I#nEed#YouR#hElP#bUYiNg#bTc#tONigHt!     "

# Eliminar los espacios al inicio y al final del texto
clean_message <- trimws(message)

# Reemplazar los "#" por espacios
clean_message <- gsub("#"," ",clean_message)

# Poner todos los caracteres en minuscula
clean_message <- tolower(clean_message)

print(clean_message)
````

### Ejercicio 2

A John le encantan los secretos y por eso ha escondido sus claves de acceso a sus bitcoins en alguna parte. Para no olvidar dónde las guardó, creó un acertijo que solo los programadores de **R** pueden descifrar. Dejó una frase escrita que dice: 

*To my lovely wife I leave my key in the safe. Behind the statue of Dio you have a picture of us both.*

Después dejó algunas instrucciones para decodificarla:

* A la longitud le restas 71, y luego la repites nuevamente pero que 72 se convierta en 69.
* Lo que está entre 34 y 35 es la segunda palabra.
* Si miras desde 43 hasta 38 no serás alguien **negativo**.

¿Dónde están escondidas las claves? 

*Pista: Utiliza indexación y slicing con la función `substr`.*

````r
phrase <- "To my lovely wife I leave my key in the safe. Behind the statue of Dio you have a picture of us both."

# Resolver primera instruccion
t1 <- substr(phrase, nchar(phrase) - 71, nchar(phrase) - 69)

# Resolver segunda instruccion
t2 <- substr(phrase, 34, 35)

# Resolver tercera instruccion
t3 <- substr(phrase, nchar(phrase) - 43, nchar(phrase) - 38)

print(paste(t1, t2, t3, sep = " "))
````

## Creación, acceso y modificación de vectores atómicos y listas

Los vectores atómicos y las listas son estructuras de datos que permiten almacenar múltiples elementos dentro de una misma variable, manteniendo un orden específico y facilitando su acceso y modificación. Estas estructuras son ampliamente utilizadas para organizar colecciones de datos y realizar operaciones repetitivas sobre conjuntos de información. En **R**, los vectores atómicos facilitan el agrupamiento de múltiples valores siempre y cuando éstos sean del mismo tipo; mientras las listas ofrecen mayor versatilidad, ya que pueden contener diferentes tipos de datos y permiten agregar, eliminar, recorrer y transformar elementos de manera dinámica. Su comprensión es fundamental para el desarrollo de algoritmos y programas más complejos como los que realizarás más adelante en tu aprendizaje.
