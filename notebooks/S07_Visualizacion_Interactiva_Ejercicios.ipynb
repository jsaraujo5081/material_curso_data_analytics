### Ejercicio 1

Carga las librerías **Tidyverse** y **plotly**.

````r
library(tidyverse)
library(plotly)
````

Crea un dataset con 10,000 observaciones aleatorias de las siguientes cuatro distribuciones estadísticas:

| Distribucion | Parametros | Función a utilizar |
| ---- | ---- | ---- |
| Normal Estandar $(Z)$ | media = 0; desviación estandar = 1 | `rnorm` |
| T de Student | grados de libertad = 5 | `rt` |
| Chi cuadrado $(\chi^2)$ | grados de libertad = 5 | `rchisq` |
| Fisher | grados de libertad = (10,10) | `rf` |

````r
df <- tibble(
    normal_estandar = rnorm(10000),
    t_student = rt(10000, 5),
    chi_cuadrado = rchisq(10000, 5),
    fisher = rf(10000, 10, 10),
)

df %>% sample_n(10)
````

Usando la función `plot_ly` genera un histograma de los datos generados con distribucion normal estandar y guárdalo en `fig_base`. Asegúrate que este gráfico de forma mínima cumpla con lo siguiente:

* El argumento `type` sea el correspondiente.
* Las barras deben ser de un color de relleno cualquiera (color A) y una transparencia al 50%. 
* Los bordes de las barras deben ser de otro color complementario (color B).

````r
# Generar histograma de distribucion normal
fig_base <- plot_ly(
    type = "histogram",
    x = df$normal_estandar,
    marker = list(
        color = "dodgerblue", 
        line = list(color = "#063e75", width = 1),
        opacity = 0.5
    ),
    width = 1000,
    height = 400
)

fig_base
````

### Ejercicio 3

Ya tienes la idea básica para crear botones, en los que es necesario especificar los argumentos a actualizar en el gráfico al hacer click en ellos. Prueba entonces tus habilidades incorporando un nuevo *widget* en `fig_base` solo que ahora el mismo debe contar con lo siguiente:

* Un primer botón que genere un histograma de los datos con color A de relleno, color B para bordes, y transparencia del 50%.
* Un segundo botón que genere un cajón con bigotes de los datos con color B para puntos (outliers), y color A para los bordes y lineas.   

````r
# Definir widget de botones
b_type <- list(

    type = "buttons",
    y = 0.8,
    buttons = list(

        # Boton para histograma
        list(
            method = "update",
            label = "Histograma",
            args = list(
                list(
                    type = "histogram",
                    marker = list(
                        color = "dodgerblue",
                        line = list(color = "#063e75", width = 1),
                        opacity = 0.5
                    )
                )
            )
        ),

        # Boton para cajon
        list(
            method = "update",
            label = "Cajón con Bigotes",
            args = list(
                list(
                    type = "box",
                    marker = list(color = "#063e75"),
                    line = list(color = "dodgerblue", width = 2)
                )
            )
        )

    )

)
````

````r
# Visualizar grafico con widget
fig_base %>% 
layout(updatemenus = list(b_type))
````

Si todo ha sido hecho correctamente, notarás que los botones funcionan adecuadamente aunque el eje "y" en el gráfico de cajón es un poco irrelevante.  En vista que esto no es un argumento de diseño en la visualización sino que es una propiedad de la funcion `layout`. Incorpora en la definición del *widget*, dentro de la lista de argumentos de cada botón, una nueva sublista que contenga el código:

```r
# Para el boton del histograma
list(yaxis = list(visible = TRUE))

# Para el boton del cajón
list(yaxis = list(visible = FALSE)
```

````r
b_type <- list(

    type = "buttons",
    y = 0.8,
    buttons = list(

        # Boton para histograma
        list(
            method = "update",
            label = "Histograma",
            args = list(
                list(
                    type = "histogram",
                    marker = list(
                        color = "dodgerblue",
                        line = list(color = "#063e75", width = 1),
                        opacity = 0.5
                    )
                ),
                list(yaxis = list(visible = TRUE))
            )
        ),

        # Boton para cajon
        list(
            method = "update",
            label = "Cajón con Bigotes",
            args = list(
                list(
                    type = "box",
                    marker = list(color = "#063e75"),
                    line = list(color = "dodgerblue", width = 2)
                ),
                list(yaxis = list(visible = FALSE))
            )
        )

    )

)
````

Verifica el funcionamiento de tu gráfico interactivo con este ajuste.

````r
fig_base %>% 
layout(updatemenus = list(b_type))
````

### Ejercicio 4

De forma muy similar a la creación de botones, es posible también incorporar un controlador a modo de menú desplegable cambiando al elemento `type` de la lista a "list" (en botones recuerda que el valor es "buttons"). 

Visto esto, define un menú desplegable para actualizar la visualización, de modo que sea posible para el usuario seleccionar cualquiera de las distribuciones de la tabla y que en `fig_base` están especificados en el argumento `x`. 

````r
m_dist <- list(

    type = "list",
    y = 0.5,
    buttons = list(

        list(
            method = "update",
            label = "Normal Estandar",
            args = list(
                list(
                    x = list(df$normal_estandar)
                )
            )
        ),

        list(
            method = "update",
            label = "T - Student",
            args = list(
                list(
                    x = list(df$t_student)
                )
            )
        ),

        list(
            method = "update",
            label = "Chi Cuadrado",
            args = list(
                list(
                    x = list(df$chi_cuadrado)
                )
            )
        ),

        list(
            method = "update",
            label = "Fisher",
            args = list(
                list(
                    x = list(df$fisher)
                )
            )
        )

    )
)
````

Muestra el gráfico resultante incorporando tu widget en la lista de `updatemenus`, y manteniendo los botones antes creados. Guarda esta visualización en la variable `fig_distribuciones`.

````r
# Actualizar visualizacion
fig_distribuciones <- fig_base %>% 
layout(
    updatemenus = list(b_type, m_dist)
)

fig_distribuciones
````

### Ejercicio 1

Si recuerdas de ejercicios previos, habíamos desarrollado una demostración visual para la **Ley de los grandes numeros** aplicada a la media muestral de unos datos simulados. Repite nuevamente los pasos allí descritos tal que:

1. Se genere un vector atómico `poblacion` con 2,000 datos aleatorios entre 0 y 100.
2. Se calcule la media poblacional de estos datos y se guarde el resultado en la variable `prom_poblacion`.
3. Se simulen muestras de estas población con tamaños de 5 a 1,000 (en saltos de 5), y en cada una se calcule y guarde la media muestral obtenida.
4. Se genere un dataframe `df_promedios` de 200 filas con dos columnas: tamaño de muestra ("n") y promedio muestral ("promedio").  

````r
# Generar poblacion simulada
poblacion <- sample(0:100, size = 2000, replace = TRUE)

# Calcular media poblacional
prom_poblacion <- mean(poblacion)

#Simular muestras y calcular sus promedios
promedios <- c()
tam_muestra <- 1:200 * 5
for (n in tam_muestra) {
    muestra <- sample(poblacion, size = n)
    promedios <- c(promedios, mean(muestra))
}

# Generar dataframe
df_promedios <- tibble(
    n = tam_muestra,
    promedio = promedios
)

glimpse(df_promedios)
````

Replica el código a continuación e interpreta lo que hace con `df_promedios`.

```r
df_promedios_f <- tibble(
    n = as.double(),
    promedio = as.double(),
    frame = as.integer()
)

for (frame in 1:nrow(df_promedios)) {
    
    df_temp <- df_promedios[1:frame,] %>%
    mutate(frame = frame)

    df_promedios_f <- df_promedios_f %>%
    bind_rows(df_temp)

}
```

````r
# Replicar codigo
df_promedios_f <- tibble(
    n = as.double(),
    promedio = as.double(),
    frame = as.integer()
)

for (frame in 1:nrow(df_promedios)) {
    
    df_temp <- df_promedios[1:frame,] %>%
    mutate(frame = frame)

    df_promedios_f <- df_promedios_f %>%
    bind_rows(df_temp)

}
````

````r
# Mirar primeras 20 filas del nuevo dataset
df_promedios_f %>%
slice_head(n = 20)
````

````r
# Verificar las dimensiones de nuevo dataset
dim(df_promedios_f)
````

Si comprendiste bien el código anterior, habrás notado que se ha "extendido" el dataset original de modo que para cada observación se replican todas las observaciones anteriores. Este proceso nos permitirá construir nuestra visualización animada en los siguientes pasos.

### Ejercicio 2

En primer lugar, y con la tabla `df_promedios`, genera un gráfico de linea con `plot_ly` que muestre la evolución del promedio muestral a medida que avanzan las iteraciones. Llama a esta gráfica `fig_base`.

````r
fig_base <- plot_ly(
    type = "scatter",
    x = df_promedios$n,
    y = df_promedios$promedio,
    mode = "lines",
    showlegend = FALSE,
    width = 700,
    height = 400
)

fig_base
````

Incorpora en este gráfico una linea horizontal de referencia para el promedio poblacional, a fin de verificar la convergencia postulada por la Ley de los grandes números.

````r
fig_base <- fig_base %>%
layout(
    shapes = list(
        list(
            type = "line",
            x0 = 0, x1 = 1, xref = "paper",
            y0 = prom_poblacion, y1 = prom_poblacion, yref = "y",
            line = list (color = "red", dash = "dot")
        )
    )
)

fig_base
````

### Ejercicio 3

Crea ahora una nueva visualización llamada `fig_lgn` y tomando como referencia lo realizado en el ejercicio precedente. Modifica en este nueva visualización lo siguiente:

* Cambia los datos fuente a la tabla `df_promedios_f`.
* Dentro de la función `plot_ly` adiciona el argumento `frame = df_promedios_f$frame`.

````r
fig_lgn <- plot_ly(
    type = "scatter",
    x = df_promedios_f$n,
    y = df_promedios_f$promedio,
    frame = df_promedios_f$frame,
    mode = "lines",
    showlegend = FALSE,
    width = 1000,
    height = 500
) %>%
layout(
    shapes = list(
        list(
            type = "line",
            x0 = 0, x1 = 1, xref = "paper",
            y0 = prom_poblacion, y1 = prom_poblacion, yref = "y",
            line = list (color = "red", dash = "dot")
        )
    )
)
````

Guarda la visualización como *lgn.html* y abrela en el navegador para ver lo que hiciste. Siéntete libre de dar click en "Play".

El argumento `frame` le indica a **plotly** que esta gráfica creada debe actualizarse automáticamente en función de la variable allí asignada, y por tanto, aquí es donde se define la animación que buscábamos.

Guarda nuevamente la visualización y ábrela para verificar si aumenta la velocidad.
