## Preparación de datos

Ejecuta tu plan de acción de manera organizada.

```r
# Ajustar nombres de columna
names(df_pixel) <- names(df_pixel) %>%
tolower() %>%
gsub(" ", "_", .)

names(df_pixel)
```

```r
# Eliminar registros con valores perdidos
df_pixel <- df_pixel %>%
filter(!is.na(id_cliente))

print(colSums(is.na(df_pixel)))
```

```r
# Extraer año, mes y dia de la semana de la fecha
df_pixel <- df_pixel %>%
mutate(
    periodo = year(fecha),
    mes = month(fecha),
    dia_semana = wday(fecha)
)

df_pixel %>% sample_n(5)
```

```r
# Hacer que los dias de semana sean mas semamnticos
df_pixel <- df_pixel %>%
mutate(dia_semana = if_else(
    dia_semana == 2,
    "01.Lun",
    if_else(
        dia_semana == 3,
        "02.Mar",
        if_else(
            dia_semana == 4,
            "03.Mie",
            if_else(
                dia_semana == 5,
                "04.Jue",
                if_else(
                    dia_semana == 6,
                    "05.Vie",
                    if_else(
                        dia_semana == 7,
                        "06.Sab",
                        "07.Dom"
                    )
                )
            )
        )
    )
))

df_pixel["dia_semana"] %>% unique()
```

```r
# Corregir observacioens erroneas en tipo_articulo
df_pixel <- df_pixel %>%
mutate(tipo_articulo = if_else(
    tipo_articulo == "ENTRAS",
    "ENTRADAS",
    tipo_articulo
))

df_pixel["tipo_articulo"] %>% unique()
```

```r
# Eliminar caracter USD de venta y volver double
df_pixel <- df_pixel %>%
mutate(valor_venta = gsub(" USD","", valor_venta)) %>%
mutate(valor_venta = as.double(valor_venta))

glimpse(df_pixel)
```

```r
# Eliminar registros duplicados
df_pixel <- df_pixel %>%
filter(!duplicated(.))

print(sum(duplicated(df_pixel)))
```

### ¿Cuál es el nivel de ventas, comensales y comandas alcanzados en el último año?

Empieza construyendo una tabla base con las siguientes columnas: "VENTAS (USD)", "# CLIENTES" y "# COMANADAS".

```r
# Crear tabla base
df_resumen <- df_pixel %>%
filter(periodo == max(periodo)) %>%
summarise(
    `VENTAS (USD)` = sum(valor_venta),
    `# CLIENTES` = n_distinct(id_cliente),
    `# COMANDAS` = n_distinct(id_orden)
)

print(df_resumen)
```

Con la información de esta tabla y mediante la función `plot_ly` crea una visualización interactiva de la misma. Ten en cuenta los siguientes argumentos mínimos:

* `type = "table"`.
* `header` que debe ser una lista nombrada (diccionario) con los nombres que desees en tu tabla interactiva.
* `cells`que debe ser una lista nombrada con los valores que deseas incluir en tu tabla interactiva. 

Usa también la función `layout` de **plotly** para ajustar los márgenes de la visualización.

```r
# Generar visualizacion interactiva
plot_ly(
    type = "table",
    width = 500,
    height = 175,
    columnwidth = c(50,50),
    header = list(
        values = NA,
        fill = list(color = "dodgerblue")
    ),
    cells = list(
        values = rbind(paste("<b>",names(df_resumen)), unlist(df_resumen[1,])),
        fill = list(color = c("lightgray", "white")),
        font = list(family = "Arial", size = c(15, 20), color = "#063e75"),
        format = list(NULL, ",.0f")
    )
) %>%
layout(
    margin = list(t = 25, r = 25, b = 25, l = 25)
)
```

### ¿Cómo se comparan estas ventas con los años previos?

Con `ggplot`, crea un gráfico de barras que muestre la evolución por año de las ventas del restaurante y guárdalo en una variable.

```r
# Crear visualizacion base
df_venta_per <- df_pixel %>%
group_by(periodo) %>%
summarise(venta = sum(valor_venta)) %>%
ungroup()

options(scipen = 999)
plt_venta_per <- ggplot(
    data = df_venta_per,
    mapping = aes(
        x = periodo,
        y = venta,
        text = paste(
            "<b>Venta (USD): </b>",
            formatC(venta, format = "f", digits = 0, big.mark = ",")
        )
    )
) + 
geom_col(
    fill = "dodgerblue",
    color = "#063e75"
) +
scale_x_continuous(
    breaks = min(df_venta_per$periodo):max(df_venta_per$periodo)
) +
labs(
    x = "",
    y = ""
) +
theme_classic()
```

Has que este gráfico de barras sea interactivo mediante la función `ggplotly`. Para esto considera el argumento `p` que debe contener el gráfico base creado antes.

```r
# Generar visualización interactiva
ply_venta_per <- ggplotly(
    p = plt_venta_per,
    tooltip = "text",
    width = 600,
    height = 400
)

ply_venta_per
```

### ¿Cómo se ha comportado el ticket promedio por comanda a través de los años?

Genera un grafico simple de linea que muestre la evolución anual del ticket promedio (venta / comandas) en el restaurante y guárdalo en una variable. Utiliza para esto `ggplot`.

```r
# Crear visualizacion base
df_ticket_per <- df_pixel %>%
group_by(periodo) %>%
summarise(ticket = sum(valor_venta) / n_distinct(id_orden)) %>%
ungroup()

plt_ticket_per <- ggplot(
    data = df_ticket_per,
    mapping = aes(
        x = periodo,
        y = ticket,
        group = 1,
        text = paste(
            "<b>Ticket (USD): </b>",
            formatC(ticket, format = "f", digits = 2, big.mark = ",")
        )
    )
) + 
geom_line(
    color = "#063e75",
    linetype = "dotted"
) +
geom_point(
    color = "dodgerblue"
) +
scale_x_continuous(
    breaks = min(df_ticket_per$periodo):max(df_ticket_per$periodo)
) +
labs(
    x = "",
    y = ""
) +
theme_classic()
```

Transforma el gráfico anterior en una visualización dinámica con `ggplotly`.

```r
# Generar visualización interactiva
ply_ticket_per <- ggplotly(
    p = plt_ticket_per,
    tooltip = "text",
    width = 600,
    height = 400
)

ply_ticket_per
```

### ¿Cómo se comporta la cantidad de comandas a través de los días de la semana y años?

Repite el mismo procedimiento que en los apartados anteriores, esto es, guarda un gráfico base de `ggplot` y luego haslo interactivo mediante `ggplotly`. La visualización resultante debería mostrar en el eje x los días de la semana, y barras conexas en cada día con la información de comandas por año. 

```r
# Crear visualizacion base
df_comandas_diaper <- df_pixel %>%
group_by(periodo, dia_semana) %>%
summarise(comandas = n_distinct(id_orden)) %>%
ungroup

plt_comandas_diaper <- ggplot(
    data = df_comandas_diaper,
    mapping = aes(
        x = dia_semana,
        y = comandas,
        fill = as.character(periodo),
        text = paste(
            periodo,
            "-",
            dia_semana,
            "\n<b> # Comandas </b>:",
            comandas
        )
    )
) +
geom_col(
    position = "dodge",
    color = "gray"
) +
labs(
    x = "",
    y = "",
    fill = ""
) +
scale_fill_brewer(palette = "Blues") +
theme_classic()

# Generar visualización interactiva
ply_comandas_diaper <- ggplotly(
    p = plt_comandas_diaper,
    tooltip = "text",
    width = 800,
    height = 400
) 

ply_comandas_diaper
```

### ¿Existen diferencias en las preferencias de consumo dependiendo del punto de venta?

Repite el procedimiento para presentar un gráfico de barras apiladas que muestre para cada punto de venta la distribución por tipo de articulo consumido.

```r
# Crear visualizacion base
df_pref_pdv <- df_pixel %>%
mutate(tipo_articulo = if_else(
    tipo_articulo %in% c("MENU DEGUSTACION", "MENU INFANTIL"),
    "MENU",
    if_else(
        tipo_articulo %in% c("BEBIDAS ESPIRITUOSAS", "COCTELES"),
        "BEBIDAS ALCOHOLICAS",
        if_else(
            tipo_articulo %in% c("CERVEZAS", "VINOS"),
            "CERVEZAS Y VINOS",
            if_else(
                tipo_articulo %in% c("ENTRADAS", "PICADAS"),
                "ENTRADAS Y PICADAS",
                tipo_articulo
            )
        )
    )
)) %>%
group_by(punto_venta, tipo_articulo) %>%
summarise(comandas = n_distinct(id_orden)) %>%
ungroup() %>%
group_by(punto_venta) %>%
mutate(pct_comandas = 100 * comandas / max(cumsum(comandas))) %>%
ungroup()

plt_pref_pdv <- ggplot(
    data = df_pref_pdv,
    mapping = aes(
        x = punto_venta,
        y = pct_comandas,
        fill = tipo_articulo,
        text = paste(
            tipo_articulo,
            "\n <b> % Comandas: </b>",
            formatC(pct_comandas, format = "f", digits = 1),
            "% (",
            formatC(comandas, format = "f", digits = 0, big.mark = ","),
            ")",
            sep = ""
        )
    )
) +
geom_col(
    color = "gray",
    width = 0.5
) +
labs(
    x = "",
    y = "",
    fill = ""
) +
scale_fill_brewer(palette = "Blues") +
theme_classic() +
theme(
    axis.line.y = element_blank(),
    axis.text.y = element_blank(),
    axis.ticks.y = element_blank()
)

# Generar visualizacion interactiva
ply_pref_pdv <- ggplotly(
    p = plt_pref_pdv,
    tooltip = "text",
    width = 800,
    height = 400
) 

ply_pref_pdv
```

### ¿Qué porcentaje de los clientes que han consumido en el último año son repetidores?

Vamos a responder esta pregunta mediante un gráfico de pie. Si bien **ggplot** permite construir este tipo de visualizaciones con la función `coord_polar` (puedes investigar cómo hacerlo en este [link](https://r-graph-gallery.com/piechart-ggplot2.html)), esta función no es compatible con `ggplotly`. Ante lo expuesto, empezemos construyendo una tabla base que cuente la cantidad de clientes del último año que han tenido más de una comanda así como aquellos que solo han tenido un consumo.

```r
# Extraer clientes del ultimo año
clientes_ult <- df_pixel %>%
filter(periodo == max(periodo)) %>%
pull(id_cliente)

# Crear tabla base
df_repetidores <- df_pixel %>%
filter(id_cliente %in% clientes_ult) %>%
group_by(id_cliente) %>%
summarise(comandas = n_distinct(id_orden)) %>%
ungroup() %>%
mutate(tipo = if_else(
    comandas > 1,
    "SI regresa",
    "NO regresa"
)) %>%
count(tipo) %>%
arrange(n)
```

Con `plot_ly`construye ahora el gráfico de pie considerando los siguientes argumentos mínimos:

* `type = "pie"`.
* `labels` para especificar el tipo de comensal (repetidor o no).
* `values` para especificar la cantidad de cada uno de los tipos de comensal.

Usa también la función `layout` de **plotly** para ajustar los márgenes de la visualización y ocultar la leyenda.

```r
# Generar visualizacion interactiva
ply_repetidores <- plot_ly(
    type = "pie",
    width = 400,
    height = 400,
    labels = df_repetidores$tipo,
    values = df_repetidores$n,
    sort = FALSE,
    textinfo = "label+percent",
    hovertemplate = paste(
        "<b> # Clientes: </b>%{value:,.0f}",
        "<extra></extra>"
    ),
    marker = list(
        colors = c("dodgerblue", "#063e75")
    )   
) %>%
layout(
    showlegend = FALSE,
    margin = list(t = 25, r = 25, b = 25, l = 25)
)

ply_repetidores
```

### ¿Con qué frecuencia regresan estos clientes repetidores?

Genera una visualización interactiva que muestre la distribución de comensales repetidores del último año en las siguiente categorias excluyentes:

* Clientes que regresan al restaurante trimestralmente.
* Clientes que regresan semestralmente.
* Clientes que regresan anualmente.
* Clientes que regresan de forma bi - anual.
* Clientes que regresan de forma tri - anual.
* Cliente que regresan de forma esporádica.

```r
# Crear visualizacion base
df_frec_rep <- df_pixel %>%
filter(id_cliente %in% clientes_ult) %>%
group_by(id_cliente) %>%
summarise(comandas = n_distinct(id_orden)) %>%
ungroup() %>%
filter(comandas > 1) %>%
mutate(frec_dias = as.integer(max(df_pixel$fecha) - min(df_pixel$fecha)) / comandas) %>%
mutate(tipo_frec = if_else(
    frec_dias < 90,
    "a.Trimestral",
    if_else(
        frec_dias < 180,
        "b.Semestral",
        if_else(
            frec_dias < 360,
            "c.Anual",
            if_else(
                frec_dias < 360 * 2,
                "d.Bianual",
                if_else(
                    frec_dias < 360 * 3,
                    "e.Trianual",
                    "f.Esporadico"
                )
            )
        )
    )
)) %>%
count(tipo_frec)

plt_frec_rep <- ggplot(
    data = df_frec_rep,
    mapping = aes(
        x = tipo_frec,
        y = n,
        text = paste(
            "<b># Clientes: </b>",
            formatC(n, format = "f", digits = 0)
        )
    )
) +
geom_col(
    fill = "dodgerblue",
    color = "#063e75"
) +
coord_flip() +
labs(
    x = "",
    y = ""
) +
theme_classic()

# Generar visualizacion interactiva
ply_frec_rep <- ggplotly(
    plt_frec_rep,
    tooltip = "text",
    width = 400,
    height = 400
) 

ply_frec_rep
```
