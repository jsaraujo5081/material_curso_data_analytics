## Definición y uso de funciones

### Ejercicio 1

Escribe una función llamada `promedios` que calcule las principales medidas de centralidad de dos números cualquiera y devuelva los resultados a modo de diccionario. Para esto considerar los siguientes estadísticos:

* Media aritmética ($\bar{x}_a$):

$$ \bar{x}_a = \frac{x_1 + x_2}{2} $$

* Media geométrica ($\bar{x}_g$):

$$ \bar{x}_g = \sqrt{|x_1 \times x_2|} $$

* Media armónica ($\bar{x}_h$):

$$ \bar{x}_h = \frac{2}{\frac{1}{x_1} + \frac{1}{x_2}} = \frac{2\times x_1 \times x_2}{x_1 + x_2} $$

````r
promedios <- function (x_1, x_2) {
    med_a = (x_1 + x_2) / 2
    med_g = abs(x_1 * x_2) ** (0.5)
    med_h = (2 * x_1 * x_2) / (x_1 + x_2)
    dicc = c(
        "media aritmetica" = med_a, 
        "media geometrica" = med_g, 
        "media armonica" = med_h
    )
    return(dicc)
}
````

Comprueba la función creada con dos números cualquiera.

````r
print(promedios(5.8, 3.7))
````

### Ejercicio 2

Escribe una función llamada `varianza` la cual calcule y devuelva este estadístico ($var$) para dos números cualquiera. Para esto considerar la siguiente definición:

$$ var = \frac{(x_1 - \bar{x}_a)^2 + (x_2 - \bar{x}_a)^2}{2} $$

Recuerda que $\bar{x}_a$ es la media aritmética.

````r
varianza <- function (x_1, x_2) {
    med_a = (x_1 + x_2) / 2
    var = ((x_1 - med_a) ** 2 + (x_2 - med_a) ** 2) / 2
    return(var)
}
````

Comprueba la función creada con dos de números cualquiera. Guarda tu resultado en una variable llamada `var`.

````r
var <- varianza(20.3, 12.1)
print(var)
````

Obtén la raíz cuadrada de `var` y guárdala en una variable llamada `desv`. Este resultado se conoce como desviación estándar o típica.

````r
desv <- var ** (0.5)
print(desv)
````

## Creación y uso de estructuras condicionales

### Ejercicio 1

Escribe una función llamada `es_bisiesto` que, dado un año `a` como entrada, devuelva si es bisiesto o no. Recuerda que para saber si un año es bisiesto, se debe cumplir con las siguientes condiciones:

* El año debe ser divisible por 4.
* Si es divisible por 100, pero NO es divisible para 1000, entonces NO es bisiesto (i.e. 2000 es bisiesto pero 2100 no lo es).

````r
es_bisiesto <- function (a) {
    bisiesto = TRUE
    # Probar primera condicion
    if (a %% 4 == 0) {
        # Probar segunda condicion
        if (a %% 100 == 0 & a %% 1000 != 0) {
            bisiesto = FALSE
        } 
    } else {
        bisiesto = FALSE
    }
    return(bisiesto)
}
````

Comprueba si los siguientes años son bisiestos: 96, 100, 1492, 1886, 1990, 2000, 4100.

````r
# Crear vector atomico con los años solicitados
periodos <- c(96, 100, 1492, 1886, 1990, 2000, 4100)

# Generar bucle para comprobar años con funcion
for (p in periodos) {
    print(paste("El año", p, "es bisiesto:", es_bisiesto(p)))
}
````

### Ejercicio 2

Escribe una función `es_primo` que reciba como entrada un número natural cualquiera e imprima un mensaje donde indique si el número es primo o compuesto. Vale indicar que un número primo es aquel que solamente es divisible por 1 o por sí mismo, por ejemplo el 7, 11, o 13.

````r
es_primo <- function (n) {

    primo = TRUE

    if (n > 2) { # 1 y 2 por defecto son primos entonces no se los considera
        for (d in 2:(n-1)) {
            if (n %% d == 0) {
                primo = FALSE
                stop # El comando stop hace parar al bucle
            }
        }
    }

    if (primo) {
        print(paste(n, "es primo"))
    } else {
        print(paste(n, "es compuesto"))
    }

}
````

Verifica que la función creada funciona para los primeros 50 números enteros. ¿Notas algún patrón?

````r
for (n in 1:50) {
    es_primo(n)
}
````

La sucesión de Fibonacci es un conjunto de números naturales muy famoso que deberías conocerlo al menos por cultura general. En esta sucesión se cumple que el número $a$ en la posición $i$ está dado por $a_i = a_{i-1} + a_{i_2}$, con $a_1 = 0$ y $a_2 = 1$. En este sentido, los primeros elementos de esta sucesión son:

$$ 0, 1, 1, 2, 3, 5, 8, 13, ...$$

Utiliza funciones y condicionales para calcular rápidamente cuál es el 40-número de Fibonacci.


````r
# Definir funcion
fibonacci <- function (n) {
  a1 = 0
  a2 = 1
  if (n == 1) {
    resultado = a1
  } else if (n == 2) {
    resultado = a2
  } else {
    resultado = fibonacci (n - 1) + fibonacci(n - 2) # Se pueden llamar a funciones en si mismas
  }
  return(resultado)
}

# Calcular y mostrar numero 40 de Fibonacci
print(fibonacci(40))
````
