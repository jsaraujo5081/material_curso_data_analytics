### Ejercicio 5

Descifra el siguiente mensaje en código usando funciones de **R**:

*sgn1rts@3d@0su@l3@n3@0tr3px3@nu@rar3d1sn0c@s3d3up@3t@ay*

````r
mensaje <- "sgn1rts@3d@0su@l3@n3@0tr3px3@nu@rar3d1sn0c@s3d3up@3t@ay"

#Reemplazar @ por espacios
mensaje <- gsub("@", " ", mensaje)

#Reemplazar 3 por e
mensaje <- gsub("3", "e", mensaje)

#Reemplazar 1 por i
mensaje <- gsub("1", "i", mensaje)

#Reemplazar 0 por o
mensaje <- gsub("0", "o", mensaje)

#Invertir el mensaje
mensaje <- paste(rev(strsplit(mensaje, NULL)[[1]]), collapse = "")

print(mensaje)
````

## Creación, acceso y modificación de vectores atómicos y listas

### Ejercicio 1

Crea un vector atómico `years` que contenga los años entre 2014 y 2025, e imprime su longitud.

````r
years <- c(2014:2025)
print(length(years))
````

Crea un vector atómico `dividendos` que contenga 12 números reales cualquiera entre 1 y 3, y que representan los dividendos de la empresa DataCorp en los años generados en la lista anterior. Imprime el tipo de variable de `dividendos`.

````r
dividendos <- c(1.2, 1.35, 1.5, 1.6, 1.8, 1.95, 2, 2.1, 2.3, 2.5, 1.98, 2.31)
print(typeof(dividendos))
````

### Ejercicio 2

Calcula la diferencia en los dividendos entre el primer y el último año del vector, y guarda el resultado en una variable llamada `diferencia`.

````r
diferencia <- dividendos[length(dividendos)] - dividendos[1]
````

Crea un nuevo vector atómico llamado `dividendos_ultimos` que contenga solo los dividendos de los últimos 5 años.

````r
dividendos_ultimos <- dividendos[(length(dividendos) - 4): length(dividendos)]
````

Calcula el promedio simple de los dividendos en los últimos 5 años usando `dividendos_ultimos`. Recuerda que el promedio simple $\mu$ se obtiene de la fórmula:

$$\mu = \frac{\sum x}{n}$$

donde $x$ son los elementos a promediar y $n$ es la cantidad de elementos.

````r
promedio <- sum(dividendos_ultimos) / length(dividendos_ultimos)
print(promedio)
````

### Ejercicio 3

Calcula la tasa acumulada de crecimiento de los `dividendos` desde 2014 hasta 2025. Guarda el resultado en la variable `tacc`.

````r
tacc <- diferencia / dividendos[1]
````

Usando el `tacc`, obtén el crecimiento porcentual promedio `p` usando la siguiente fórmula:

$$p = \sqrt[n-1]{1+tacc} - 1$$

Imprime el resultado obtenido.

````r
p <- (1 + tacc) ** (1 / (length(dividendos) - 1)) - 1
print(paste(formatC(p * 100, format = "f", digits = 1), "%", sep = ""))
````

Guarda la información de años, dividendos y crecimiento porcentual promedio en una lista.

````r
info_dividendos <- list(
    years = years,
    dividends = dividendos,
    var = p
)
info_dividendos
````
