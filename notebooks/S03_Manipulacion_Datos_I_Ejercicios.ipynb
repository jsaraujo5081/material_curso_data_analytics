## Tratamiento de valores ausentes

### Ejercicio 1

Has un conteo de los valores ausentes que existen en cada una de las columnas del dataset.

```r
print(colSums(is.na(df)))
```

### Ejercicio 2

Imputa en los valores ausentes de la columna VehicleTipe el valor modal. 

```r
# Calcular la moda
moda <- df %>%
count(VehicleType) %>%
arrange(desc(n)) %>%
slice_head(n = 1) %>%
pull(VehicleType)

# Asignar moda a valores perdidos
df <- df %>%
mutate(VehicleType = if_else(
    is.na(VehicleType),
    moda,
    VehicleType
))

print(colSums(is.na(df)))
```

### Ejercicio 3

Imputa en los valores ausentes del resto de columnas la expresión "not defined".

```r
df <- df %>%
mutate(Gearbox = if_else(
    is.na(Gearbox),
    "not defined",
    Gearbox
)) %>%
mutate(Model = if_else(
    is.na(Model),
    "not defined",
    Model
)) %>%
mutate(FuelType = if_else(
    is.na(FuelType),
    "not defined",
    FuelType
)) %>%
mutate(NotRepaired = if_else(
    is.na(NotRepaired),
    "not defined",
    NotRepaired
))

print(colSums(is.na(df)))
```

## Agrupamiento de datos

### Ejercicio 1

Muestra el precio promedio, máximo y mínimo para cada tipo de vehículo usando `group_by`.

```r
df %>%
group_by(VehicleType) %>%
summarise(
    mean = mean(Price),
    max = max(Price),
    min = min(Price)
) %>%
ungroup()
```

### Ejercicio 2

Muestra en una sola tabla resumen el precio promedio, el kilometraje mediano y la cantidad de modelos por marca.

```r
df %>%
group_by(Brand) %>%
summarise(
    price_mean = mean(Price),
    km_med = median(Kilometer),
    q_models = n_distinct(Model)
) %>%
ungroup()
```
